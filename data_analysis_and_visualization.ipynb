```python
# Import the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load your data
df = pd.read_csv('housing_price_data.csv')

# Preview the data
print(df.head())
```

```python
# Summary statistics
print(df.describe())
```

```python
# You can see the max price above is 329,999
# It is very absurd price for rental. It might be a mistake
# For safety measure, just remove all entry above 50,000.

df_cleaned = df[df['price'] < 50000] # Upper limit at 50,000

# Verify that the entry is gone
print(df_cleaned.describe())
```

```python
# Replace the dataframe with the cleaned one
df = df_cleaned
```

# Exploratory Data Analysis (EDA)

## Distribution of Prices

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True)
plt.title('Distribution of Property Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

## Distribution of Sizes

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['size'], kde=True)
plt.title('Distribution of Property Sizes')
plt.xlabel('Size (sq ft)')
plt.ylabel('Frequency')
plt.show()
```

## Prices vs Sizes

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='size', y='price', data=df)
plt.title('Price vs. Size')
plt.xlabel('Size (sq ft)')
plt.ylabel('Price')
plt.show()
```

## Prices vs Number of Bedrooms

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='bedroom', y='price', data=df)
plt.title('Price vs. Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```

## Prices by Location

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='location', y='price', data=df)
plt.title('Property Prices by Location')
plt.xlabel('Location')
plt.ylabel('Price')
plt.xticks(rotation=90)
plt.show()
```
